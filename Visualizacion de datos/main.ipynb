### Limpieza de columnas

```python
def clean_price_column(df, price_column):
    df[price_column] = df[price_column].astype(str)
    df[price_column] = df[price_column].apply(lambda x: re.sub(r'[^\d.]', '', x))
    df[price_column] = pd.to_numeric(df[price_column], errors='coerce')
    return df

for name, df in dataframes.items():
    if 'price' in df.columns:
        df = clean_price_column(df, 'price')
    elif 'discounted_price' in df.columns:
        df = clean_price_column(df, 'discounted_price')
        df.rename(columns={'discounted_price': 'price'}, inplace=True)
    dataframes[name] = df

print("DataFrames limpiados:")
print(dataframes.keys())
```

### Unificacion de en un solo Dataframe

```python
def clean_price_column(df, price_column):
    df[price_column] = df[price_column].astype(str)
    df[price_column] = df[price_column].apply(lambda x: re.sub(r'[^\d.]', '', x))
    df[price_column] = pd.to_numeric(df[price_column], errors='coerce')
    return df

for name, df in dataframes.items():
    df['category'] = name.replace('.csv', '')
    if 'price' in df.columns:
        df = clean_price_column(df, 'price')
    elif 'discounted_price' in df.columns:
        df = clean_price_column(df, 'discounted_price')
        df.rename(columns={'discounted_price': 'price'}, inplace=True)
    
    if 'Unnamed: 0' in df.columns:
        df.drop(columns=['Unnamed: 0'], inplace=True)
    
    dataframes[name] = df

all_products_df = pd.concat(dataframes.values(), ignore_index=True)

print("DataFrame combinado:")
print(all_products_df.head())

```

### Analizar las Categorías con los Precios Más Altos

```python
def clean_price_column(df, actual_price_column):
    df[actual_price_column] = df[actual_price_column].astype(str)
    df[actual_price_column] = df[actual_price_column].apply(lambda x: re.sub(r'[^\d.]', '', x))
    df[actual_price_column] = pd.to_numeric(df[actual_price_column], errors='coerce')
    return df

# Añadir una columna de categoría basada en el nombre del archivo y limpiar precios
for name, df in dataframes.items():
    # Añadir columna de categoría
    df['category'] = name.replace('.csv', '')
    
    # Limpiar la columna de precios si existe 'price' o 'discounted_price'
    if 'actual_price' in df.columns:
        df = clean_price_column(df, 'actual_price')
    elif 'discounted_price' in df.columns:
        df = clean_price_column(df, 'discounted_price')
        df.rename(columns={'discounted_price': 'actual_price'}, inplace=True)
    else:
        # Si no se encuentra ninguna columna de precio, mostrar un mensaje de advertencia
        print(f"No se encontró ninguna columna de precio en el DataFrame '{name}'.")
    
    # Eliminar columna 'Unnamed: 0' si existe
    if 'Unnamed: 0' in df.columns:
        df.drop(columns=['Unnamed: 0'], inplace=True)
    
    dataframes[name] = df

# Unir todos los DataFrames en uno solo
all_products_df = pd.concat(dataframes.values(), ignore_index=True)

# Verificar las columnas y los primeros registros del DataFrame combinado
print("Columnas en el DataFrame combinado:")
print(all_products_df.columns)
print("\nPrimeros registros del DataFrame combinado:")
print(all_products_df.head())

# Filtrar las filas con precios válidos si la columna 'price' existe
if 'actual_price' in all_products_df.columns:
    all_products_df = all_products_df.dropna(subset=['actual_price'])
    # Encontrar las categorías con los productos más caros
    top_categories = all_products_df.groupby('category')['actual_price'].max().sort_values(ascending=False).head(10)
    print("\nCategorías con productos más caros:")
    print(top_categories)
    
    # Visualización con seaborn y matplotlib
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
    plt.title('Categorías con Productos de Precios más Altos')
    plt.xlabel('Precio Máximo')
    plt.ylabel('Categoría')
    plt.show()
else:
    print("No se pueden encontrar las categorías con productos más caros porque la columna 'actual_price' no existe en el DataFrame.")
```

## ¿Existe una correlación entre precio y calificación en los productos?

### Convertir la columna ratings a tipo numérico

```python
all_products_df['ratings'] = pd.to_numeric(all_products_df['ratings'], errors='coerce')
```

### Filtrar filas con valores válidos en actual_price y ratings

```python
valid_df = all_products_df.dropna(subset=['actual_price', 'ratings'])
```

### Calcular la correlación

```python
correlation = valid_df['actual_price'].corr(valid_df['ratings'])
print(f"Correlación entre precio y calificación: {correlation}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x=valid_df['actual_price'], y=valid_df['ratings'])
plt.title('Relación entre Precio y Calificación')
plt.xlabel('Precio')
plt.ylabel('Calificación')
plt.show()
```

## ¿Cuál es el rango de precio dependiendo del tipo de subcategoría?

### Agrupar por subcategoría y calcular el rango de precios

```python
price_ranges = all_products_df.groupby('sub_category')['actual_price'].agg(['min', 'max'])

price_ranges = price_ranges.sort_values(by='max', ascending=False)

plt.figure(figsize=(12, 10))
sns.set_style('whitegrid')

sns.barplot(x=price_ranges.index, y=price_ranges['max'], color='skyblue', label='Precio Máximo')
sns.barplot(x=price_ranges.index, y=price_ranges['min'], color='lightgreen', label='Precio Mínimo')

plt.xlabel('Subcategoría')
plt.ylabel('Precio')
plt.title('Rango de Precios por Subcategoría')
plt.legend()

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```
